### 다음 뉴스 테이터 추출
- 뉴스 기사에서 제목, 작성자, 작성일, 댓글 갯수 추출
- id의 경우 원칙적으로 한 html문서 내에 유일
- id, class 속성으로 찾기

```python
# bs4모듈에 정의된 BeautifulSoup클래스를 사용한다
from bs4 import BeautifulSoup
# 예시 html문서
mydoc = '''
<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>샘플페이지</title>
</head>
<body>
  <p>안녕하세요.BeautifulSoup</p>
</body>
</html>
'''
# html 파싱
soup = BeautifulSoup(mydoc, "html.parser")

# 페이지 제목 추출, 출력
print(soup.title.string)
```

### CSS를 이용하여 tag찾기
- select, select_one 함수 사용
- css selector 사용법
	- 태그명 찾기
	- 자식 태그 찾기
	- 아이디 찾기 #id
	- 클래스 찾기 .class
	- 속성값 찾기[name='test']
		- 속성값 prefix 찾기[name='test']
		- 속성값 prefix 찾기[name='test']
	- n번째 자식 tag찾기 : nth-child(n)

```python
#h3 태그가 있는  정보 모두
import requests
url = "https://v.daum.net/v/20190728165812603"
res = requests.get(url)
soup = BeautifulSoup(res.text)
soup.select("h3")
```

```python
# 부분 문자열인데 t로 시작하는 것 가져오세요.
soup.select("h3[class ^='t']")
```

```python
# 기자 이름과 입력 시간을 찾아보시오.
soup.select("span.txt_info")
```

```python
soup.select('span.txt_info:nth-child(1)')
```

```python
soup.select('span.txt_info')[0]
```

```python
soup.select('span.txt_info')[0].get_text()
```
