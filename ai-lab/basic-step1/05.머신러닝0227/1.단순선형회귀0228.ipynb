```python
# 그래프 한글 처리를 위한 환경 설정
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm # 한글 폰트 관련
import matplotlib
from matplotlib import rc
path = "../data/malgun.ttf"
font_name = fm.FontProperties(fname=path).get_name()
matplotlib.rc("font", family = font_name)
plt.rc("font", family="Malgun Gothic")
```

```python
#데이터 생성 : y = 2x+1
np.random.seed(42)
# 0~2사이의 100개 데이터
X = 2*np.random.rand(100,1)
y = 2 * X + 1 + np.random.randn(100,1) # 선형관계 + 노이즈
```

```python
y
```

```python
# 데이터 시각화 - 데이터 분포도
plt.scatter(X, y, color="red", label="데이터")
plt.xlabel("독립변수 X")
plt.ylabel("종속변수 y")
plt.title("데이터 분포")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model  import LinearRegression
# 모델 학습
model = LinearRegression() # 단순 선형 회귀 알고리즘을 구현한 객체이름- 객체생성
model.fit(X, y) # 데이터 X, y를 기반으로 모델을 학습시킴
# 이 과정에서 모델은 입력 데이터에 가장 적합한 직선회귀선을 찾게 된다.
# 가장 적합한 직선 회귀선을 찾는 알고리즘이 필요함.
# 알고리즘을 어떤 모델로 하는가에 따라서 정확도가 달라 진다.
```

```python
# 새로운 데이터 생성 및 예측
X_new = np.array([[0],[2]]) # 새로운 입력 데이터. 0과2 두 개의 값을 포함 2차배열
y_predict = model.predict(X_new) #학습된 모델을 바탕으로 X_new에 해당 하는 예측 결과

plt.scatter(X, y, color="blue", label="데이터")
plt.plot(X_new, y_predict, color="red", label="예측선", linewidth=2)
plt.xlabel("독립변수 X")
plt.ylabel("종속변수 y")
plt.title("단순선형회귀 예측 결과")
plt.legend()
plt.show()
```

```python
# 학습된 파라미터 출력
# 모델을 적용한 기울기와 절편을 출력해 본다.
print(f"기울기(w) : {model.coef_[0][0]:.2f}")
print(f"절편 (b) : {model.intercept_[0]:.2f}")
```
