```python
m = folium.Map(location=[36.763718, 127.28194050], zoom_start=15, tiles='OpenStreetMap', width=800, height=450)
folium.Marker([36.763718, 127.28194050],
            #   popup="KOREATECH",
              tooltip="거래금액",
              icon=folium.Icon(color='red', icon='glyphicon-heart')).add_to(m)
m
```
