```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
""" 
그래프의 스타일을 설정하는 데 사용되는 함수입니다. Seaborn은 Matplotlib 위에서 동작하며, 
그래프의 시각적 스타일을 손쉽게 설정하고 변경할 수 있도록 다양한 스타일 옵션을 제공
set_style(): Seaborn에서 그래프의 스타일을 설정하기 위한 함수입니다. 
이 함수는 그래프의 배경, 격자선, 틱 마크 등의 시각적 요소를 제어합니다.
"whitegrid": 스타일 옵션 중 하나로, 흰색 배경에 격자선이 있는 스타일을 의미합니다.
"""
sns.set_style("whitegrid")
# 셀의 결과를 아래쪽에 그래프를 첨가해서 보여줍니다
%matplotlib inline
```

주요 스타일 옵션     
Seaborn은 다양한 스타일 옵션을 제공합니다.    
"darkgrid": 배경이 회색이고, 격자선이 있는 스타일입니다. 기본 스타일로, 데이터의 패턴을 쉽게 확인할 수 있습니다.    
"whitegrid": 배경이 흰색이고, 격자선이 있는 스타일입니다. 깔끔하고 명확한 시각화를 제공합니다.    
"dark": 회색 배경에 격자선이 없는 스타일입니다. 그래프의 데이터를 강조하고 싶을 때 사용됩니다.     
"white": 흰색 배경에 격자선이 없는 스타일입니다. 가장 간단하고 깔끔한 스타일입니다.     
"ticks": 흰색 배경에 축 틱 마크가 강조된 스타일입니다. 틱 마크가 두드러지게 표시되어 데이터 포인트를 구분하는 데 유용합니다.     

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Seaborn 스타일 설정
sns.set_style("whitegrid")

# 임의의 데이터 생성
data = np.random.normal(size=100)

# 히스토그램 그리기
sns.histplot(data)
plt.title("Histogram with Whitegrid Style")
plt.show()
```

### lmplot(linear model plot)
- 컬럼들 간의 선형 관계 확인할 수 있는 그래프 
- 회귀 분석을 포함해서 시각화 해주는 그래프로 의미가 있습니다.    
:matplotlib에는 없는 그래프
- 참고 링크 : https://seaborn.pydata.org/generated/seaborn.lmplot.html

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset("tips")
tips.head(5)
```

```python
# 전체 금액 대비 팁이 높은지를 살펴 봅시다
# y축에 종속변수 대상인 팁을 표시 합니다.
# 두 컬럼끼리 관계가 있을까를 보기 위해 그리는 그래프 입니다. 
# 선형 회귀 그래프 입니다.    
# 총금액과 팁은 관계가 있는 것으로 보입니다. 
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7)
plt.show()
```

하나로 보여지는 선형을 흡연자별 둘로 나누어 그려 주었습니다.    
마치 3차원 그래프 처럼 표현해 보았습니다.     

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, height=7)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set2", height=7)
plt.show()
```

### heat map
- heatmap : 2차원 데이터를 heatmap 형태로 시각화하는 방법      
- seaborn의 heatmap() 함수를 사용      
- 참고 링크 : https://seaborn.pydata.org/generated/seaborn.heatmap.html    
import numpy as np    
import seaborn as sns    

히트맵은 2차원 데이터를 색상으로 표현하여 시각적으로 비교하거나 패턴을 찾는 데 유용한 그래프입니다.     
sns.heatmap 함수는 다음과 같은 다양한 옵션을 제공하여 히트맵의 스타일과 색상을 조정할 수 있습니다:     
annot: True로 설정하면 각 셀에 값이 표시됩니다.     
cmap: 히트맵의 색상 팔레트를 설정합니다. 기본적으로 YlGnBu라는 색상 팔레트가 사용되지만,     
cmap 인자를 통해 다양한 색상 맵을 사용할 수 있습니다.     
linewidths: 각 셀 간의 경계선을 설정합니다. 경계선의 두께를 조절할 수 있습니다.     
vmin 및 vmax: 색상 강도의 최소값과 최대값을 설정하여 색상 범위를 조정할 수 있습니다.     

```python
uniform_data = np.random.rand(10, 12)
uniform_data.shape
```

```python
uniform_data
```

생성되는 시각화 설명  
이 코드를 실행하면 10x12 크기의 히트맵이 생성됩니다:
   
x축과 y축: 히트맵의 x축과 y축은 각각 데이터의 열과 행을 나타냅니다.    
셀의 색상: 각 셀의 색상은 해당 위치의 값에 따라 달라지며, 값이 클수록 색상이 더 진해집니다.      
데이터 패턴 확인: 히트맵은 데이터의 전반적인 패턴, 예를 들어 특정 행 또는 열에서의 값의 변화,     
고유한 데이터의 클러스터 등을 시각적으로 쉽게 파악할 수 있습니다.     

```python
sns.heatmap(uniform_data)
plt.show()
```

```python
sns.heatmap(uniform_data, vmin=0, vmax=1)
plt.show()
```

### pairplot

#### 붓꽃(Iris) 데이터 


```python
sns.set(style="ticks")
iris = sns.load_dataset("iris") # 붓꽃의 데이터  꽃잎과 꽃받침정보가 담김
iris.head()
```

```python
iris.shape
```

```python
# 4*4 그래프가 그려진다. 왼쪽 상단에서 오른쪽 하단으로 대각선을 그어서 위아래는 대칭이 되는 그래프이다.
# 그래서 한쪽  면만 살펴 보면 됩니다. 직선에 가까운 그래프가 상관관계가 높다고 말합니다. 
# x값이 커질때 y값도 커지면 양의 상관관계가 높다고 말합니다. 
# 상관관계가 가장 높을 때가 1 입니다. -1을 가지면 음의 상관관계가 높다고 말합니다.
# 결론 : 직선에 가까울 수록 상관관계가 높다고 말합니다.
plt.figure(figsize=(4,4))
sns.pairplot(iris)
plt.show()
```

```python
#군집화 공부할 때도 그려 봄니다. 
sns.pairplot(iris, hue="species") # 꽃의 종류가 3가지여서 같이 보여준다.
plt.show()
```

```python
# 대각선을 그려서 위아래가 같은 그래프 인데 x축과 y축이 바뀌어 그려져서 달라보인다.
# sepal_length가 클수록 sepal_width가 관계를 갖고 있는지 보는 그래프 입니다.
# sepal_length와 sepal_width는 관계가 없어 보임니다. 
sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
plt.show()
```

```python
# 숫자형 데이터만 선택하여 상관관계 계산
# 상관관계가 1에 가까울 수록 관계가 높다.여기서는 petal_width와 petal_length가
# 0.96으로 1에 가깝다 
iris_corr = iris.drop(columns=['species']).corr()
iris_corr
```

```python
# 위 상관관계 그래프에서 관계가 높이 나온 petal_length와 petal_width만
# 따로 그래프를 그려 봄니다.
sns.pairplot(iris, vars=["petal_length", "petal_width"])
plt.show()
```
